## (1) concat 메서드로 데이터 연결하기

## 1) 행 방향으로 데이터 연결하기

- concat 메서드에 연결하려는 데이터프레임을 리스트에 담아 전달하면 연결된
  데이터프레임을 반환한다.
- concat 메서드는 데이터프레임을 연결할 때 열이름이 같으면 위에서 아래
  방향으로 연결한다. 인덱스도 그대로 유지된다.

```python
import pandas as pd

df1 = pd.read_csv('data2/concat_1.csv')
df2 = pd.read_csv('data2/concat_2.csv')
df3 = pd.read_csv('data2/concat_3.csv')

df1,df2,df3
```

```python
row_concat = pd.concat([df1,df2,df3])  #데이터프레임 아래로 연결
row_concat
```

```python
row_concat.iloc[3,] #인덱스3 의 열 추출
```

## 3) 공통 열과 공통 인덱스만 연결하기
- 열 이름이 일부 다른 세개의 데이터프레임을 행으로 연결하면
  열이름이 정렬되고, 데이터프레임에 없는 열은 누락값(NaN)으로 처리된다.

```python
#새로운 열이름 다시 지정
df1.columns = ['A','B','C','D']
df2.columns = ['E','F','G','H']
df3.columns = ['A','C','F','H']
```

```python
df1,df2,df3
```

```python
row_concat = pd.concat([df1,df2,df3],axis=0) #행방향
row_concat
```

- 데이터프레임의 공통 열만 골라 연결하면 누락값이 생기지 않는다.
(공통열인것만 골라서 연결. 아닌것 버리면)
- join = inner로 지정하면 공통 열만 골라 연결할 수 있다.
- df1,df2,df3은 공통열이 없다, df1, df3은 공통 열 A와 ,C만 연결된다.

```python
pd.concat([df1,df2,df3],join='inner') #세개 다 공통인 열이 없어서 값이안나옴
```

```python
pd.concat([df1,df3],join = 'inner')
```

```python
pd.concat([df2,df3],join='inner',ignore_index=True)
```

- 인덱스를 다시 지정후 df1,df2,df3를 axis = 1로 해서 열방향으로 연결한다.
- join = inner로 지정하면 공통 행만 골라 출력된다

```python
df1.index=[0,1,2,3]
df2.index=[4,5,6,7]
df3.index=[0,2,6,7]
df1,df2,df3
```

```python
col_concat = pd.concat([df1,df2,df3],axis=1)
```

## (2) merge로 옆으로(열방향) 통합하기
- 두데이터를 옆으로(열방향)으로 통합한다.
- 기준데이터와 정보를 결합하고 싶은 데이터간에 공통된 열에 대해
  같은 값을 찾아서 결합해 하나의 데이터로 통합한다.
- 실습을 위해 특정 위치의 날씨 정보에 필요한 데이터 집합을 불러온다
- person 관측한 사람이름, site 관측위치, visited 관측날짜, survey  날씨정보 이다.
- 관측날짜(visited) 의 일부 데이터만 추출후 사용한다.

```python
person = pd.read_csv('data2/survey_person.csv')
site = pd.read_csv('data2/survey_site.csv')
survey = pd.read_csv('data2/survey_survey.csv')
visited = pd.read_csv('data2/survey_visited.csv')

person,site,survey,visited
```

```python
visited_subset = visited.loc[[0,2,6],]
visited_subset
```

```python
#name 이랑 site 열을 기준(같은것)으로 visited와 site 데이터를 통합함
o2o_merge = site.merge(visited_subset,left_on='name',right_on='site')
o2o_merge
```

```python
m2o_merge = site.merge(visited,left_on='name',right_on='site')
m2o_merge
```

```python
#person과 survey데이터 통합
ps = person.merge(survey,left_on='ident',right_on='person')
ps
```

```python
#visited와 survey데이터 통합
vs = visited.merge(survey,left_on='ident',right_on='taken')
print(vs)
```

- left_on, right_on에 여러개 열을 리스트에 담아서 전달할 수 있다.
- ps데이터프레임에 ident, taken,quant, reading 열의 값과  vs 데이트프레임에
  person, ident, quant, reading 열의 값을 이용하여 ps 데이터프레임과
  vs 데이터프레임을 서로 연결한다.
- 양쪽 데이터프레임에 있었던 중복된 열 이름(ident,taken,person)에
  접미사 _x, _y가 추가되어 있다.
- _x는 왼쪽, _y는 오른쪽 데이터프레임의 열을 의미한다.

```python
#merge한걸 활용해 다시 merge
ps_vs = ps.merge(vs,left_on=['ident','taken','quant','reading'],
                right_on=['person','ident','quant','reading'])
```

```python
ps_vs
```

- merge 메서드의 how 속성을 사용하면 기준 데이터를 정할 수 있다.
- how='left' 지정하면 왼쪽 데이터프레임을 기준으로 통합하므로 오른쪽 데이터프레임에
  매칭되는 값이 없다면 NaN으로 결과가 출력된다.
- 분석을 위한 엑셀 데이터 읽어온다.

```python
sample_1 = pd.read_excel('data2/sample_1.xlsx',
                        header=1,
                        skipfooter=2,
                        usecols='A:c') #a부터c컬럼 값만 가져오고싶다
sample_1
```

```python
code_master = pd.read_excel('data2/sample_codemaster.xlsx')
code_master
```

```python
sample_1_code = pd.merge(left=sample_1,
                        right = code_master, 
                        how = 'left', #leftouterjoin같이 왼쪽데이터 기준 
                         #오른쪽데이터에 없는 데이터는 none으로 표시.
                        left_on = '국적코드',
                        right_on='국적코드')
```

```python
sample_1_code
```

```python
sample_1_code = pd.merge(left=sample_1,
                        right = code_master, 
                        how = 'inner', #서로 있는것만 
                        left_on = '국적코드',
                        right_on='국적코드')
sample_1_code
```

## (3) append 메서드로 아래로(행방향)통합하기
- 2019년 11월 입국객수를 가지고있는 sample_1_code의 기준년도열을 만든다.

```python
sample_1_code['기준년월'] = '2019-11'
sample_1_code
```

```python
#sample_1과 아래로 통합할 2를 만드는중
sample_2 = pd.read_excel('data2/sample_2.xlsx',
                        header=1,
                        skipfooter=2,
                        usecols='A:C')

sample_2['기준년월'] = '2019-12'
```

```python
#국적코드와 열방향 통합(옆으로)
sample_2_code = pd.merge(left=sample_2,
                        right=code_master,
                        how='left',
                        left_on='국적코드',
                        right_on='국적코드')
sample_2_code
```

- 두데이터는 컬럼이 똑같기때문에 아래로 (행방향) 통합할 수 있다.
- ignore_index = True 인덱스값을 다시 새롭게 설정한다. 

```python
sample = sample_1_code.append(sample_2_code,ignore_index=True)
sample
```

```python
sample.to_excel('data2/sample.xlsx',index=False) # 인덱스번호 제외하고 저장
```

## 2) 누락값 처리 - 변경, 삭제
- 누락값은 누락값을 임의의 값으로 변경하거나 데이터프레임의 이미 존재하는
  값으로 대신 채우는 방법 등으로 처리할 수 있다.

- 데이터프레임의 fillna 메서드에 0을 대입하면 누락값을 0으로 변경한다.
- 데이터프레임의 크기가 크고 메로리를 효율적으로 사용해야 하는 경우에 
  자주 사용한다.

```python
ebola.fillna(0).iloc[0:10,0:5]
```

- fillna(method='ffill')로 지정하면 누락값의 앞행의 값으로 채워진다.
  3행의 누락값은 2행의 값으로 대체한다. 0,1행은 처음부터 누락값이기 때문에
  그대로 NaN값이 된다.


```python
ebola.fillna(method='ffill').iloc[0:10,0:5]
```

```python
#뒤의값으로 채워주기
ebola.fillna(method='bfill').iloc[0:10,0:5]
```

```python
#누락값 앞뒤의 행 값을 이용해 중간값을 구함
ebola.interpolate().iloc[0:10,0:5]
```

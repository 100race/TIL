## 시세종목 업비트 API로 코인목록 가져오기
https://docs.upbit.com/reference#%EC%A3%BC%EB%AC%B8-%EA%B0%80%EB%8A%A5-%EC%A0%95%EB%B3%B4

```python
#시세종목조회 API로 받아오기
import requests

url = "https://api.upbit.com/v1/market/all"

querystring = {"isDetails":"false"}

headers = {"Accept": "application/json"}

response = requests.request("GET", url, headers=headers, params=querystring)

print(response.text)
```

## 코인테이블에 데이터입력

```python
sql = "INSERT INTO COIN(eng,kor) VALUES (%s, %s);"
```

```python
#response 데이터 정제 후 DB에 저장
coin_check = [] #중복확인용 리스트

for x in response.text.split('{'):
    #print(x)
    if(len(x)>=3): #쓰레기값 제거
        eng= x.split(':')[1].split(',')[0].split('-')[1].split('"')[0] #코인 영문 컬럼
        kor = x.split(':')[2].split(',')[0].split('"')[1] #코인 한글명 컬럼
        #중복확인후 저장
        if eng not in coin_check:
            coin_check.append(eng) 
            cursor.execute(sql,[eng,kor])
            print([eng,kor])
```

```python
coin_db.commit() #커밋 필수
```

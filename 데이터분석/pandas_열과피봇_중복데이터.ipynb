# 1. 열과 피벗
- 데이터프레임의 열 자체가 어떤값을 의미할때가 있다.
  그러다 보니 데이터프레임의 열이 옆으로 길게 늘어선 형태가 된다.
- 이런 데이터를 넓은 데이터라 한다. 
- 판다스는 데이터프레임을 깔끔한 데이터로 정리하는데 유용한
  melt 메서드를 제공한다.

## (1) melt 메서드 사용하기

1) 1개의 열만 고정하고 나머지 열을 행으로 바꾸기
- 퓨 리서치 센터(pew Research Center)에서 조사한 '미국의 소득과 종교' 라는 데이터를
  사용한다. 11개 열이 있는 데이터프레임이다. 소득정보가 열을 구성하고 있다.

```python

import pandas as pd
pew = pd.read_csv('data2/pew.csv')
pew.head()
```

```python
pew.iloc[:,0:6]
```

- 소득정보 열들을 행 데이터로 옮긴다 id_vars = 'religion'으로 지정한다.
  religion만 제외한 나머진 소득정보 열들이  variable 열로 정리되고 소득 정보의 
  행데이터가 value열로 정리되었다.
- 이것을 ' religion 열을 고정하여 피벗했다 ' 라고 말한다.

※ 피벗이란 회전하는 물체의 균형을 잡아 주는 중심점이라는 뜻이다.
※ 피벗 했다는 것은 여러 데이터 중에서 자신이 원하는 데이터만을 가지고 원하는
   행과 열에 데이터를 배치하여 새로운 데이터 집합을 만드는 것이다.

```python
#id_vars로 하나의 고정할 열이름 지정, 나머지는 variable,value열로 재배치
pew_long= pd.melt(pew,id_vars='religion')
pew_long.head()
```

```python
#자동 생성되는 variable, value이름을 var_name, value_name 인자값 사용
pew_long= pd.melt(pew,id_vars='religion',var_name ='income',value_name='count')
pew_long.head()
```

```python
billboard = pd.read_csv('data2/billboard.csv')
billboard.iloc[0:5,0:16]
```

```python
#2개 이상의 열을 고정하고 나머지 열을 행으로 바꾸기
#- year, artist, track,time, date.entered 열 고정, 나머지 열(wk1,wk2......)을 피벗한다.
billboard_long = pd.melt(billboard,
                        id_vars=['year','artist','track','time','date.entered'],
                        var_name='week',value_name='rating')
billboard_long
```

## (2) pivot_table 메서드 사용하기
 - pivot_table이란 기존 데이터의 칼럼을 재구성해서 데이터에 대한 통계를 한눈에
   파악할 수 있게 정리한 표의 형태이다.
 - 데이터통합 데이터를 sample.xlsx파일로 저장했는데 실습을 위해 다시 읽어온다.

```python
sample = pd.read_excel('data2/sample.xlsx')
sample
```

- 국적과 기준월에 따른 입국객수의 평균값을 추출한다.
 - pivot_table 메서드의 인자값 설명 확인하기
 - values -> 피벗테이블의 내부에 들어갈 값 설정
   index -> 피벗테이블의 행으로 배치할 값 설정
   columns -> 피벗테이블의 열로 배치할 값 설정
   aggfunc  -> values값을 index와 columns기준으로 평균값을 계산한다.
 - aggfunc의 값으로 mean(평균), sum(합계), min(최솟값), median(중앙값),max(최댓값),
   count(개수) nunique(중복을 제거한 후 개수,원소개수) 등을 사용한다.


```python
sample_pivot = sample.pivot_table(values='입국객수',#셀에보여줄값
                                 index='국적명', #행으로배치
                                 columns='기준년월', #열로 배치
                                 aggfunc='mean') #values의 평균값
sample_pivot
```

- index, column는 항상 있어야 하는 것이 아니다. 필요에 의해 index만 배치할 수도 있다.
- 국적에 따른 최대 입국객 수를 알아본다.

```python
sample_pivot_2 = sample.pivot_table(values='입국객수',
                                   index='국적명',
                                   aggfunc='max')
sample_pivot_2
```

## (3) melt , pivot_table 메서드 사용하기
- 기상 데이터의 날짜열(d1,...d31)에 월별 최고, 최저 온도 데이터가 저장되어 있다.
- 날짜 열이 옆으로 길게 늘어져 있어 보기 불편하다.
- melt 메서드로 일별 온도(d1,d2,...)를 피벗하여 day열에 날짜 열이 정리되고
  날짜 열의 데이터는 temp열에 정리된다.

```python
weather = pd.read_csv('data2/weather.csv')
weather.iloc[:5,:11]
```

```python
weather_melt = pd.melt(weather,
                      id_vars=['id','year','month','element'], #고정할값
                      var_name='day',
                      value_name='temp')
weather_melt.head()
```

- 월별 최고,최저 온도를 한눈에 보여주기 위해 pivot_table 메서드를 사용한다.
- pivot_table은 행과 열의 위치를 다시 바꿔 정리를 해준다.
- index 인자는 위치를 그래도 유지할 열 이름을 지정한다.
- columns 인자는 피벗할 열 이름을 지정한다.
- values 인자는 새로운 열의 데이터가 될 이름을 지정한다.

```python
weather_tidy = weather_melt.pivot_table( values = 'temp',
                                        index=['id','year','month','day'],
                                        columns='element'  
)
weather_tidy
```

```python
weather_tidy_flat = weather_tidy.reset_index() #인덱스를 새로지정
weather_tidy_flat.head()
```

  중복됩니다. 
- 일관성 유지가 필요한 가수(artist)는 고유한 값으로 따로 관리하는것이 좋다.
- 분석중간에 반복되는 artist 이름이 변경되면 분석내용이 잘못될 수 있다.

```python
#노래제목이 loser인 데이터검색
billboard_long[billboard_long.track=='Loser'].head()
```

```python
#중복데이터를 가지고있는 열을 따로 모아 새로운 데이터프레임에 저장
billboard_songs = billboard_long[['year','artist','track','time']]
billboard_songs.shape
```

```python
#drop_duplicates메서드로 데이터프레임의 중복데이터를 제거한다
billboard_songs = billboard_songs.drop_duplicates()
billboard_songs.shape
```

```python
billboard_songs = billboard_songs.drop_duplicates()
billboard_songs.shape
```

```python
#중복한 데이터프레임에 아이디 열을 추가한다
billboard_songs['id'] = range(len(billboard_songs))
billboard_songs.head(n=10)
```

- merge 메서드로 노래정보와 주간 순위 데이터를 합친다.
- 같은 노래에 대한 주별 차트 진입에 대한 데이터를 좀더 깔끔하고
  보기 좋게 확인할 수 있다.

```python
billboard_ratings = billboard_long.merge(
billboard_songs,
on=['year','artist','track','time'])

billboard_ratings.shape
```
